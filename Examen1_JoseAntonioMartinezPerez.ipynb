```python
import numpy as np
import scipy as sp
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
e5_1 = '{:.6f}'.format(math.e**-5)
for j in range(30):
    if '{:.6f}'.format(sum([((-1)**i * 5**i)/np.math.factorial(i) for i in range(j)]))[0:7] == e5_1[0:7]:
        print('Valor mínimo de n:', j)
        break
        
e5_1 = '{:.6f}'.format(math.e**-5)
for j in range(1,30):
    if '{:.6f}'.format(1/sum([5**i / np.math.factorial(i) for i in range(j)]))[0:7] == e5_1[0:7]:
        print('Valor mínimo de n:', j)
        break
```

```python
#1:
Ea = np.fabs(np.pi-(22/7))
print('Error absoluto 1: {:.7f}'.format(Ea))
Er = Ea/np.fabs(np.pi)
print('Error relativo 1: {:.7f}'.format(Er))

#2:
Ea = np.fabs(math.e**10 - 22000)
print('\nError absoluto 2: {:.7f}'.format(Ea))
Er = Ea/np.fabs(math.e**10)
print('Error relativo 2: {:.7f}'.format(Er))

#3:
Ea = np.fabs(10**np.pi - 1400)
print('\nError absoluto 3: {:.7f}'.format(Ea))
Er = Ea/np.fabs(10**np.pi)
print('Error relativo 3: {:.7f}'.format(Er))

#4:
Ea = np.fabs(np.math.factorial(9) - ((18*np.pi)**0.5 * (9/math.e)**9))
print('\nError absoluto 4: {:.7f}'.format(Ea))
Er = Ea/np.fabs(np.math.factorial(9))
print('Error relativo 4: {:.7f}'.format(Er))
```

```python
import sympy as sym

x = sym.Symbol('x')
print(sym.diff( 1 - math.e**x + (math.e - 1)*sym.sin((sym.pi*x)/2)))

print('\n', sym.diff((x - 1)*sym.tan(x) + x*sym.sin(sym.pi*x)))

print('\n', sym.diff(x*sym.sin(sym.pi*x)-(x-2)*sym.ln(x)))

print('\n', sym.diff((x - 2)*sym.sin(x)*sym.ln(x + 2)))
```

```python
f1 = lambda x2: -1.0*(math.e**x2) + 0.859140914229523*np.pi*np.cos(np.pi*x2/2)

f2 = lambda x2: np.pi*x2*np.cos(np.pi*x2) + (x2 - 1)*(np.tan(x2)**2 + 1) + np.sin(np.pi*x2) + np.tan(x2)

f3 = lambda x2:np.pi*x2*np.cos(np.pi*x2) - np.log(x2) + np.sin(np.pi*x2) + (2 - x2)/x2

f4 = lambda x2:(x2 - 2)*np.log(x2 + 2)*np.cos(x2) + (x2 - 2)*np.sin(x2)/(x2 + 2) + np.log(x2 + 2)*np.sin(x2)

x2 = np.linspace(-5,5,1000)

plt.plot(x2, f1(x2))
plt.title('a.')
plt.grid()
plt.show()

plt.plot(x2, f2(x2))
plt.title('b.')
plt.grid()
plt.show()

plt.plot(x2, f3(x2))
plt.title('c.')
plt.grid()
plt.show()

plt.plot(x2, f4(x2))
plt.title('d.')
plt.grid()
plt.show()
```
